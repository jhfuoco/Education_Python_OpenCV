cv2.destroyAllWindows()

## shape와 Type

채널(Channel)
- 컬러: RGB -> 3채널
- 흑백: 1채널

모든 채널 : 0~255 값  
255 : 흰색, 0: 검정색

이미지 : 배경색(하얀색)

````python
import cv2
img = cv2.imread("/Users/jihoon/venvs/workspace/python_opencv/data/panda.jpg") # 해당경로 이미지 읽어오기

print(type(img)) #Numpy Type
print(img.shape) #세로, 가로, Channel
````

읽기 옵션

1. cv2.IMREAD_COLOR : 컬러이미지, 투명영역은 무시(기본값)
2. cv2.IMREAD_GRAYSCALE : 흑백이미지
3. cv2.IMREAD_UNCHANGED : 투명영역(Alpha)까지 포함

RGB : RED, GREEN, BLUE  
RGBA : RED, GREEN, BLUE, ALPHA(투명도, 0.0 ~ 1.0) -> (255,255,255,0.5) 반투명 하얀색

````python
import cv2

img_color = cv2.imread("/Users/jihoon/venvs/workspace/python_opencv/data/panda.jpg", cv2.IMREAD_COLOR)
img_gray = cv2.imread("/Users/jihoon/venvs/workspace/python_opencv/data/panda.jpg", cv2.IMREAD_GRAYSCALE)
img_unchanged = cv2.imread("/Users/jihoon/venvs/workspace/python_opencv/data/panda.jpg", cv2.IMREAD_UNCHANGED)

print(f"shape: color -> {img_color.shape}")
print(f"shape: gray -> {img_gray.shape}")
print(f"shape: unchanged -> {img_unchanged.shape}")

cv2.imshow("color", img_color)
cv2.imshow("gray", img_gray)
cv2.imshow("unchanged", img_unchanged)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

## matplotlib 라이브러리 사용해서 출력

cv2.imshow : OpenCV 제공 => Pycharm, VScode  
plt.imshow : Matplotlib 제공 => Jupyter, Colab

````python
import cv2
import matplotlib.pyplot as plt

img_color = cv2.imread("/Users/jihoon/venvs/workspace/python_opencv/data/panda.jpg", cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.show()

img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB))
plt.show()
````
