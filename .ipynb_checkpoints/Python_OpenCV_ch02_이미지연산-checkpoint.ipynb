## 1. OpenCV를 활용한 이미지 크기 및 픽셀 정보 확인

````python
import cv2

# 상대경로 : 현재 파일 위치를 기준으로 찾기
# 절대경로 : 처음부터 찾아 들어가기

img = cv2.imread("./data/panda.jpg")

print(img.shape) # 이미지 크기(세로, 가로, 채널)
print(img.size) # 픽셀 수 확인

# BGR 순서로 출력
px = img[100,100] # Numpy 1x3 배열(세로, 가로, 채널)->[100:세로, 100:가로]
print(f'세로 100, 가로 100번째 픽셀의 BGR값 추출: {px}')

print(f'픽셀에서 RED값만 추출 {px[2]}') # R값
````

## OpenCV를 활용한 특정 범위 픽셀 변경

````python
import cv2
import matplotlib.pyplot as plt
import time

img = cv2.imread("./data/panda.jpg")

# 1. 중첩 for문을 활용한 기본적인 방법
start_time = time.time()
for i in range(0,200):
    for j in range(0,200):
        img[i,j] = [255,255,255]
print(f'seconds:{time.time()-start_time}')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
````

````python
img = cv2.imread("./data/panda.jpg")

# 2. 슬라이싱 연산을 사용한 방법(효율성, 속도 좋음)
start_time = time.time()
img[0:200, 0:200] = [0,0,0]
print(f'seconds:{time.time()-start_time}')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
````

## OpenCV를 활용한 ROI(Region of Interest: 관심영역) 추출

ROI: 특정 이미지에서 어떤 유의미한 특징을 추출할 때 특징이 들어있는 공간ㅠ

````python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./data/panda.jpg")

# Numpy Slicing : ROI 처리 가능
roi = img[100:300, 250:550] # 관심영역 추출

# ROI 단위로 이미지 복사하기
img[0:200, 0:300] = roi # 해당위치에 관심영역 넣기(추출한 관심영역과 크기 같아야함)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
````

````python
## OpenCV를 활용한 픽셀별 색상 다루기

import cv2
import matplotlib.pyplot as  lt

img = cv2.imread("./data/panda.jpg")
img[:,:,2] = 0 #RED를 0으로 변경(모든 픽셀에서 RED 제거->B,G만 남음)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
````
